```python
def stdDevOfLengths(L):
    """
    L: a list of strings

    returns: float, the standard deviation of the lengths of the strings,
      or NaN if L is empty.
    """
    if len(L) == 0:
      return float('NaN')
    mean = sum([len(s) for s in L]) / len(L)
    var = sum([(len(s) - mean)**2 for s in L]) / len(L)
    return var**0.5
```

```python
def coefVar(L):
    """Calculates the coefficient of variation of the numbers in the list L,
    using the formula coefVar = sigma / mu

    Args:
        L (list of ints): a list of ints
    """
    mu = sum(L) / len(L)
    sigma = (sum([(i - mu)**2 for i in L]) / len(L)) ** 0.5
    return sigma / mu
```

Finger Exercise: Write a code for a Monte Carlo simulation which evaluates the integral

$\int_{0}^{\pi} \sin(x) dx$

```python
import random
import math
from matplotlib import pyplot as plt

def myintegral(numDraws):
    dots_in_area = 0
    for draw in range(numDraws):
        x = random.uniform(0.0, math.pi)
        y = random.uniform(0.0, 1.0)
        if y < math.sin(x):
            dots_in_area += 1
    return (dots_in_area / numDraws) * math.pi

myintegral(1000)
```

```python
def myintegralSim(numTrials, drawPerTrial):
    results = []
    for trial in range(numTrials):
        results.append(myintegral(drawPerTrial))
    return results
        
drawPerTrial = 1000 
fig, axs = plt.subplots(3, 1, sharex=True, figsize=(6, 6))
for idx, numTrials in enumerate([100, 1000, 10000]):
    results = myintegralSim(numTrials, drawPerTrial)
    mu = sum(results) / len(results)
    sigma = (sum([(result - mu)**2 for result in results]) / len(results)) ** 0.5
    axs[idx].hist(results, bins=200, density=True)
    axs[idx].set(ylabel="Density")
    axs[idx].annotate(f"{numTrials} trials\n$\mu =$ {mu:.2f}\n$\sigma =$ {sigma:.3f}", xy=(0.05, 0.95), xycoords='axes fraction', fontsize=10, verticalalignment='top')
    # axs[idx].legend()
axs[0].set(title="Distribution of the simulated value of definite integral $\int_{0}^{\pi} \sin(x) dx$ with Monte Carlo")
axs[2].set(xlabel="Definite integral value")
```

```python
def noReplacementSimulation(numTrials):
    '''
    Runs numTrials trials of a Monte Carlo simulation
    of drawing 3 balls out of a bucket containing
    3 red and 3 green balls. Balls are not replaced once
    drawn. Returns the a decimal - the fraction of times 3 
    balls of the same color were drawn.
    '''
    numSameColor = 0
    for trial in range(numTrials):
        hand = []
        bucket = ["r"] * 3 + ["b"] * 3
        for draw in range(3):
            hand.append(bucket.pop(random.randint(0, len(bucket) - 1)))
        if hand == ["r"] * 3 or hand == ["b"] * 3:
            numSameColor += 1
    return numSameColor / numTrials
```
