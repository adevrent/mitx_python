If the probability of moving from i to j in one time step is Pr(j|i) = Pi,j, the stochastic matrix P is given by using Pi,j as the i-th row and j-th column element

$\
r_{i,j}(n)=\sum^m_{k=1}r_{i,k}(n-1) p_{k,i}$

```python
import numpy as np

def calculate_r(P, n):
    """
    P: right stochastic matrix (values in each row add up to 1).
    n: number of transitions.
    """
    if n == 0:
        return np.eye(P.shape[0])
    elif n == 1:
        return P
    else:
        return np.dot(calculate_r(P, n-1), P)
```

```python
calculate_r(np.array([[0.5, 0.5, 0], [0.2, 0.3, 0.5], []]), 5)
```
