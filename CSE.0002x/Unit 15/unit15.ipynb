$x_ f = \frac{V_0^2}{g}\sin 2\theta _0$\
\
where\
\
$x_ f$: impact location \
$V_0^2$: initial velocity \
$g$: gravitational acceleration constant \
$\theta_0$: angle between the starting velocity's path and the ground

We are interested in the effect of randomness in both $V_0$ and $\theta_0$ on \
the impact location.

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng()

Xmu = 0  # expected value (mean)
Xsigma = 1  # standard deviation
shape = (1000000,)  # length of array
X = rng.normal(Xmu, Xsigma, shape)

plt.hist(X, bins=200, density=True)
```
