Recall the canonical form of an IVP:

$\large\frac{{\rm d}\underline{u}}{{\rm d}t} = \underline{f}(\underline{u}, t)$

For the case of a linear system,  $\underline{f}$ will be a purely linear function of $\underline{u}$ and so can be written in the following form:\
$\large\frac{{\rm d}\underline{u}}{{\rm d}t} = A\underline{u}+ \underline{b}(t)$\
\
where $A$ is an $M$ x $M$ matrix and does NOT depend on time. The length $M$ vector $\underline{b}(t)$ is a function of time, but not $\underline{u}$

**Equilibrium Condition:**\
We wish to determine if a steady solution exists. (i.e. ${\rm d}\underline{u}/{\rm d}t = 0$)\
For the linear system case, the equation that must be solved to determine $\underline{u}_{\rm eq}$ is:\
\
$\large\frac{{\rm d}\underline{u}_{\rm eq}}{{\rm d}t} = A\underline{u}_{\rm eq} + \underline{b}_{\rm eq} = 0 \Rightarrow A\underline{u}_{\rm eq} = -\underline{b}_{\rm eq}$
